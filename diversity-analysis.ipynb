```r
library(tidyverse)

diversity <- read.csv("diversity_school.csv") %>%
    mutate(category = as.factor(category)) %>% 
    filter(total_enrollment > 55000) %>% 
    select(-state) %>% 
    subset(category != "Unknown" & category != "Women" & category != "Total Minority" & category != "Two Or More Races")

diversity
```

```r
options(repr.plot.width = 12, repr.plot.height = 10)

diversity_plot <- diversity %>% 
    ggplot(aes(fill = category, x = name, y = enrollment)) +
    geom_bar(position="stack", stat = "identity") +
    labs(x = "University name", y = "Number of students enrolled", fill = "Race", title = "Number of Students enrolled by Race") +
    theme(text = element_text(size = 20)) +
    theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1))

diversity_plot
```
